```python
import pandas as pd
import numpy as np
import datetime as dt
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('buys.csv')
data.head(10)
```

```python
data.describe()
```

```python
data.info()
```

```python
data.hist(figsize=(20, 12))
```

```python
data['total_buy']=data['price']*data['size']
```

```python
data.head()
```

```python
data['updated_at2'] = data['updated_at'].apply(lambda x: pd.to_datetime(x))
```

```python
data.info()
```

```python
data.head()
```

```python
pin_date = max(data['updated_at2']) + dt.timedelta(1)
```

```python
rfm = data.groupby('user').agg({
    'updated_at2': lambda x: (pin_date - x.max()).days,
    'order': 'count',
    'total_buy': 'sum'
})
rfm
```

```python
data[data['user'] == 2065.0]
```

```python
rfm.rename(columns= {
    'updated_at2': 'Recency',
    'order': 'Frequency',
    'total_buy': 'Monetary'
}, inplace=True)
rfm
```

```python
Recency_Range=[rfm['Recency'].min(),rfm['Recency'].max()]
Recency_Range
```

```python
Frequency_Range=[rfm['Frequency'].min(),rfm['Frequency'].max()]
Frequency_Range
```

```python
Monetary_Range=[rfm['Monetary'].min(),rfm['Monetary'].max()]
Monetary_Range
```

```python
r_labels = range(4, 0, -1)
r_groups = pd.qcut(rfm['Recency'], q=4, labels=r_labels)
f_labels = range(1, 5)
f_groups = pd.qcut(rfm['Frequency'], q=4, labels=f_labels)
m_labels = range(1, 5)
m_groups = pd.qcut(rfm['Monetary'], q=4, labels=m_labels)
```

```python
rfm['R'] = r_groups.values
rfm['F'] = f_groups.values
rfm['M'] = m_groups.values
rfm
```

```python
X = rfm[['R', 'F', 'M']]
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300)
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
rfm['kmeans_cluster'] = kmeans.labels_
```
